# Use LSTMs model to predict stock price trend
- we cannot predict the exactly stock price, because the new thing is independent to the past
- use 2012-2016 open_stock data to predict 2017's first_month open_stock_price

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from keras.wrappers.scikit_learn import KerasClassifier
from keras.preprocessing import image
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV
from sklearn.metrics import classification_report,confusion_matrix,roc_auc_score
from sklearn.preprocessing import StandardScaler,MinMaxScaler,Normalizer
from sklearn.preprocessing import PolynomialFeatures
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer
from sklearn.decomposition import PCA,KernelPCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

from keras.models import Sequential
from keras.layers import Convolution2D,MaxPooling2D,Dense,Flatten
from keras.layers import Dropout
from keras.layers import LSTM
```

```python
df=pd.read_csv("Google_Stock_Price_Train.csv")
```

```python
df["Date"]=pd.DatetimeIndex(df["Date"])
```

```python
# observe the 2012-2016 google stock price trend 
plt.figure(figsize=(16,9))
df["Open"].plot(label="open")
plt.legend()
plt.ylabel("Price")
```

```python
training_set=df.iloc[:,1].values
```

```python
# Deep Leaning 很吃 feature_scaling
scaler=MinMaxScaler(feature_range=(0,1))
training_set=scaler.fit_transform(training_set.reshape(-1,1))
```

# Data Structure Design
- Creating a Data Structure wuth 60 timesteps and 1 output
- RNN view each time t has 60 seq_dim(stock proce) to predict the next output
- stock operate about 20 days in a month, we view 3 months(60 days) as input_sequence model
    - ex: The 60th stock price, we use index[0:59]'s open_stock to predict

```python
# 60 timestamps input
X_train=[]

# 1 output
y_train=[]

# y_train use upper bound
for i in range(60,len(training_set)):
    X_train.append(training_set[i-60:i,0])
    y_train.append(training_set[i,0])
```

```python
len(y_train)
```

```python
len(X_train)
```

```python
#numpy transform in order to feed into rnn model
X_train,y_train=np.array(X_train),np.array(y_train)
```

```python
y_train
```

```python
X_train
```

# Reshaping
- Reshape  to (batch_size,seq_dim,input_dim)
- keras RNN Layer 3D tensor

```python
X_train=X_train.reshape(X_train.shape[0],X_train.shape[1],1)
```

# Build RNN model
- LSTM Layer, the most important parameters : output units,input_shape,return_Sequential
- based on experience:
    - units equals to 50 is better ,
    - input_shape = (seq_dim,number of predictors) 
- Because I predict the continuous data , so the cost_function I use MSE

```python
regressor=Sequential()
```

```python
# Add first LSTM Layer
regressor.add(LSTM(units=50,return_sequences=True,input_shape=(X_train.shape[1],1))) 
regressor.add(Dropout(0.2))
```

```python
# Add Second LSTM Layer
regressor.add(LSTM(units=50,return_sequences=True)) 
regressor.add(Dropout(0.2))
```

```python
# Add Third LSTM Layer
regressor.add(LSTM(units=50,return_sequences=True)) 
regressor.add(Dropout(0.2))
```

```python
# Add The Foruth LSTM Layer
regressor.add(LSTM(units=50,return_sequences=False)) 
regressor.add(Dropout(0.2))
```

```python
# Add the output Layer
regressor.add(Dense(1,activation=None))# Linear function f(x)=x
```

```python
# stock price like regression => cost function use MSE
regressor.compile(optimizer="adam",loss="mean_squared_error") # MSE
```

# Training Model
- training model take 3d (batch_size,timestamp,number of indicators)，

```python
regressor.fit(X_train,y_train,epochs=50,batch_size=24)
```

# Predict and Evaluate by test_data
- Step1.read testing data
- Step2.fetch intersted data from training for evaluate，
- Step3.Struct Data Structure for Sequence
- Step4.Transform format in order to feed into LSTM/RNN
- Step5.Predict，do Inverse_func ，I want to know the real stock_price, not the transformed data

```python
# Actual Data
test=pd.read_csv("Google_Stock_Price_Test.csv",usecols=["Open"])
```

```python
# Get Predict Data
input_=pd.concat([df["Open"],test["Open"]],axis=0,ignore_index=True)

input_ = input_[len(input_)-len(test)-60:].values 

# scale data
input_ = scaler.transform(input_.reshape(-1,1))
```

```python
input_data=[input_[i-60:i] for i in range(60,80)]
input_data=np.array(input_data)
input_data=input_data.reshape(len(test),60,1)
```

```python
# predict
result=regressor.predict(input_data) 
```

```python
result_real=scaler.inverse_transform(result)
```

```python
graph=test.join(pd.DataFrame(result_real))
graph.columns=["Actual Open","Predict_Open"]
graph.plot(figsize=(12,9))
```
